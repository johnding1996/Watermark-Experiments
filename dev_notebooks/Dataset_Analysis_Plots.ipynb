```python
from dev import load_json
from wordcloud import WordCloud
import matplotlib.pyplot as plt


for dataset_name in ["diffusiondb", "mscoco", "dalle3"]:
    print(dataset_name)
    prompts_json_path = (
        f"/fs/nexus-projects/HuangWM/datasets/main/{dataset_name}/prompts.json"
    )
    data = load_json(prompts_json_path)
    text = "".join(data.values())

    wordcloud = WordCloud(
        width=600, height=400, max_font_size=80, background_color="white"
    ).generate(text)
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```
